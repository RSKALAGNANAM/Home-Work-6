## Generate Cities List

```python
# List for holding lat_lngs and cities
# We want to pick cities such that we get a good distribution based on Latitude
# Therefore, I have defined 8 sections of Latitude; each one 20 degrees apart except when we go 
# further North or South towards the poles
# I am letting the program pick up as many as 200 stations within each of these ranges

# Define 8 lists to hold eight separate Latitude/Longitude 

lat_lngs_1 = []
lat_lngs_2 = []
lat_lngs_3 = []
lat_lngs_4 = []
lat_lngs_5 = []
lat_lngs_6 = []
lat_lngs_7 = []
lat_lngs_8 = []
lat_lngs_9 = []
lat_lngs_10 = []
lat_lngs_11 = []
lat_lngs_12 = []
cities = []

# Create a set of random lat and lng combinations

lats_1 = np.random.uniform(low=-90.000, high=-60.000, size=200)
lats_2 = np.random.uniform(low=-60.000, high=-40.000, size=200)
lats_3 = np.random.uniform(low=-40.000, high=-20.000, size=200)
lats_4 = np.random.uniform(low=-20.000, high=-10.000, size=200)
lats_5 = np.random.uniform(low=-10.000, high=-5.000, size=200)
lats_6 = np.random.uniform(low=-5.000, high=0.000, size=200)
lats_7 = np.random.uniform(low= 0.000, high=5.000, size=200)
lats_8 = np.random.uniform(low= 5.000, high=10.000, size=200)
lats_9 = np.random.uniform(low= 10.000, high=20.000, size=200)
lats_10 = np.random.uniform(low= 20.00, high=40.000, size=200)
lats_11 = np.random.uniform(low= 40.00, high=60.000, size=200)
lats_12 = np.random.uniform(low= 60.00, high=90.000, size=200)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

# Zip the eight sets of Latitude/Longitude pairs; these will be used for locating cities

lat_lngs_1 = zip(lats_1, lngs)
lat_lngs_2 = zip(lats_2, lngs)
lat_lngs_3 = zip(lats_3, lngs)
lat_lngs_4 = zip(lats_4, lngs)
lat_lngs_5 = zip(lats_5, lngs)
lat_lngs_6 = zip(lats_6, lngs)
lat_lngs_7 = zip(lats_7, lngs)
lat_lngs_8 = zip(lats_8, lngs)
lat_lngs_9 = zip(lats_9, lngs)
lat_lngs_10 = zip(lats_10, lngs)
lat_lngs_11 = zip(lats_11, lngs)
lat_lngs_12 = zip(lats_12, lngs)

# Identify nearest city for each lat, lng combination for each of the 8 sets defined above
# and add the city to the list of cities provided the city is not already part of it
# The risk of doing this is sometimes we may eliminate another city with the same name even though
# it is located elsewhere in the world (example, St. Petersburg in Florida and Russia) unless these 
# are spelled "slightly" differently

for lat_lng in lat_lngs_1:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

for lat_lng in lat_lngs_2:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
for lat_lng in lat_lngs_3:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

for lat_lng in lat_lngs_4:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
for lat_lng in lat_lngs_5:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

for lat_lng in lat_lngs_6:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
for lat_lng in lat_lngs_7:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
for lat_lng in lat_lngs_8:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

for lat_lng in lat_lngs_9:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

for lat_lng in lat_lngs_10:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
for lat_lng in lat_lngs_11:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

for lat_lng in lat_lngs_12:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print("Number of cities picked randomly = " + str(len(cities)))

```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set up a Dataframe to collect data only for the fields as required. 
# These are city, cloudiiness, country, date, humidity, lat, lng, max temp and wind speed

city = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

#Define "settings" to pass on the api_key through
settings = {"units":"imperial","appid": api_key}

#Set up a loop to go through and fetch the data
# Make sure to have code for exception handling

# Also keep a count of how many cities did not have data
city_count_without_data = 0

for i in range(len(cities)):
    print(f"Processing Record {i+1} for: {cities[i]}")
    try:
        city_weather_response = owm.get_current(cities[i], **settings)
    except:
        print(f"Looks like Data does not exist for: {cities[i]}.")
        city_count_without_data = city_count_without_data + 1
        continue
    city.append(city_weather_response["name"])
    cloudiness.append(city_weather_response["clouds"]["all"])
    country.append(city_weather_response["sys"]["country"])
    date.append(city_weather_response["dt"])
    humidity.append(city_weather_response["main"]["humidity"])
    lat.append(city_weather_response["coord"]["lat"])
    lng.append(city_weather_response["coord"]["lon"])
    max_temp.append(city_weather_response["main"]["temp_max"])
    wind_speed.append(city_weather_response["wind"]["speed"])
   
print("\nNumber of cities that did not have data =" + str(city_count_without_data))
    



```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Now that we have read the data, define a dataframe to hold the data
# Call the dataframe "my_weather_df"

my_weather_df = pd.DataFrame({"City":city, "Cloudiness":cloudiness, "Country":country, "Date":date, "Humidity":humidity,"Lat":lat,"Lng":lng,"Max Temp":max_temp,"Wind Speed":wind_speed})

#Clean the DataFrame - I believe this is a good habit to form

my_weather_df.dropna(how='any')

# Get the number of records after dropping

length_after_dropna = len(my_weather_df["City"])

print("Number of records after doing a dropna() = " + str(length_after_dropna)+".")

# During one of the runs, I found that the value of himidity for a city (Iquitos) was 319 percent
# We know this is an invalid entry because the maximum is 100 percent.
# Similarly, cloudiness that is reported as percent also cannot have values above 100 percent.
# Therefore we need to delete Rows where either one of these is above 100 percent.
# Also, there are places on the Earth where temperatures can be as high as 120 to 130 degrees F
# India had such temperatures this summer

# Keep only those Rows where Humidity <= 100 percent; Cloudiness <= 100 percent and Max Temp <= 140 deg F

my_weather_df = my_weather_df.loc[my_weather_df['Humidity'] <= 100]

my_weather_df = my_weather_df.loc[my_weather_df['Cloudiness'] <= 100]

my_weather_df = my_weather_df.loc[my_weather_df['Max Temp'] <= 140]

# Get the length of the Data Frame

length_after_dropping_invalid_data = len(my_weather_df["City"])

print("Number of records after dropping invalid data = " + str(length_after_dropping_invalid_data)+".")

records_dropped = length_after_dropna - length_after_dropping_invalid_data

print("Number of records dropped due to invalid data = " + str(records_dropped)+".")


# Export the file as a csv

my_weather_df.to_csv("cities.csv", index = None, header=True)

#Display the header

my_weather_df.head()
```
